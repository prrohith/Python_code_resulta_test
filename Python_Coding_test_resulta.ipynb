```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner
import random # library for random number generation

import json   
# tranforming json file into a pandas dataframe library
from pandas.io.json import json_normalize

print('Folium installed')
print('Libraries imported.')
```

```python
start_date='2020-01-12'
end_date='2020-01-19'
league='NFL'
url_scoreboard = 'https://delivery.chalk247.com/scoreboard/{}/{}/{}.json?api_key=74db8efa2a6db279393b433d97c2bc843f8e32b0'.format(league,start_date,end_date)
url_scoreboard
```

```python
results_score = requests.get(url_scoreboard).json()
```

```python
df=pd.read_json (url_scoreboard)
```

```python
daterange = pd.date_range(start_date, end_date)
df_res={}
for date in daterange:
    dt=date.strftime("%Y-%m-%d")
    if bool(df['results'][dt]):
        df_res[dt]=df['results'][dt]['data']
    else:
        print("No sport data on "+dt)
    
df_res.keys()
```

```python
dff = pd.DataFrame()
for key in df_res.keys():
    df_res1=df_res[key]
    dff=pd.concat([dff, pd.DataFrame(df_res1)], axis=1)
```

```python
dff.head()
```

```python
dff=dff.transpose()
dff.head()
```

```python
url_teamrankings='https://delivery.chalk247.com/team_rankings/NFL.json?api_key=74db8efa2a6db279393b433d97c2bc843f8e32b0'
results_rank = requests.get(url_teamrankings).json()
```

```python
teams = results_rank['results']['data']

# tranform teams into a dataframe
dataframe = json_normalize(teams)
dataframe.head()
```

```python
df_scoreboard=dff[['event_id', 'event_date', 'away_team_id','away_nick_name', 'away_city','home_team_id','home_nick_name', 'home_city']]
```

```python
df_scoreboard.head()
```

```python
df_scoreboard[['event_date','event_time']] = df_scoreboard.event_date.str.split(" ",1,expand=True,)
```

```python
mid = df_scoreboard['event_time']
df_scoreboard.drop(labels=['event_time'], axis=1,inplace = True)
df_scoreboard.insert(2, 'event_time', mid)
df_scoreboard.head()
```

```python
dataframe.rename(columns={'team_id':'away_team_id'},inplace=True)
```

```python
dataframe.head()
```

```python
team_final=pd.merge(df_scoreboard, dataframe, on='away_team_id')
team_final.head()
```

## Document says to consider the 'rank' column of team_rankings as the value for response label 'away_rank' & 'home_rank'
## It also says to consider the 'adjusted_points' column of team_rankings as the value for response label 'away_rank_points' & 'home_rank_points'

```python
team_final.drop(columns=['last_week','points','modifier','team'], inplace = True)
team_final
```

### formatting columns

```python
team_final['event_date'] = pd.to_datetime(team_final['event_date'])
team_final['event_date'] = team_final['event_date'].dt.strftime('%d-%m-%Y')
team_final["adjusted_points"] = team_final.adjusted_points.astype(float)
team_final["adjusted_points"] = team_final.adjusted_points.round(2)
```

```python
team_final['home_rank']=team_final['rank']
team_final['home_rank_points']=team_final['adjusted_points']
```

```python
team_final.rename(columns={'rank':'away_rank','adjusted_points':'away_rank_points'},inplace=True)
```

### reordering the columns

```python
ACME_final=team_final[['event_id', 'event_date', 'event_time', 'away_team_id','away_nick_name', 'away_city','away_rank', 'away_rank_points','home_team_id', 'home_nick_name','home_city','home_rank','home_rank_points']]
```
